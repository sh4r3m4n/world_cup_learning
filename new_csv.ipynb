```python
import pandas as pd
import numpy as np
```

```python
matches = pd.read_csv('./WorldCupMatches.csv')
```

```python
matches
```

```python
matches = matches.rename(index=str, columns={"Year": "year", "Home Team Name" : "team1", "Away Team Name": "team2",
                                            "Home Team Goals": "score1", "Away Team Goals": "score2"})
matches['id'] = matches.index
matches = matches.loc[:, ['id', 'year', 'team1', 'score1', 'score2', 'team2']]
```

```python
matches = matches[matches.year >= 1950]
```

```python
# The given CSV was bad
matches.loc[:, ['team1']] = matches['team1'].map(lambda name: name.replace('rn">', ''))
matches.loc[:, ['team2']] = matches['team2'].map(lambda name: name.replace('rn">', ''))
```

```python
# unicode problenms, i suspect related to ivory coast
matches.loc[:, ['team1']] = matches['team1'].map(lambda name: 'Ivory Coast' if 'Ivoire' in name else name)
matches.loc[:, ['team2']] = matches['team2'].map(lambda name: 'Ivory Coast' if 'Ivoire' in name else name)
```

```python
matches
```

```python
new_teams = set(list(matches.team1) + list(matches.team2))
```

```python
old_csv = pd.read_csv('./raw_matches.csv')
old_teams = set(list(old_csv.team1) + list(old_csv.team2))
```
